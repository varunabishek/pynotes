```python
##greatest number from a user inputted array
n=int(input("enter number of elements: "))
lista = []
for i in range(n):
    element = int(input("enter elements: "))
    lista.append(element)

largest = lista[0]
for num in lista:
    if num > largest:
        largest = num
print("array: ",lista)
print("largest number is: ",largest)


    
```
